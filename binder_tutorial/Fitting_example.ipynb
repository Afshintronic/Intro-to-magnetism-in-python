First off, we import `numpy`, `pyplot` and a curve fitting module from the `scipy.optimize` package

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

In this example, we'll use Euler's formula for fitting the sum of two complex exponential functions with a cosine.
$$\frac{e^{-i\omega x}+e^{i\omega x}}{2}=\cos(\omega x)$$
For producing our exponential distributions, we use the first function on the cell below to generate the data, with some random noise to keep it interesting. Aditionally, I have added an offset in order to have a parameter which I can estimate. The second one is the fitting function. `a`, `b` and `omega` are free parameters to be determined, though we have good ideas of what they should be.

```python
def expEuler(x, offset):
    return np.real(0.5*(np.exp(-complex(0, 1.45)*x)+np.exp(complex(0, 1.45)*x))) + np.random.uniform(low = -0.5, high = 0.5, size = len(x)) + offset
def cos(x, a, b, omega):
    return a*np.cos(omega*x) + b
```

First, let's generate and plot the data. The input of the functions can be numbers or, more interenstingly, arrays. A linear space is an array of evenly spaced consecutive decimal numbers. For clarity, I'll assign the name x to a simple linear space of 500 steps from 0 to $2\pi$. The y function is our generated noisy sum of exponential functions.

```python
x = np.linspace(0, 2*np.pi, 500)
y = expEuler(x, 3) # OFfset is set at 3
plt.plot(x, y, '.')
```

It should come as no surprise that it looks like a noisy cosine with a frequency higher than 1. To estimate the parameters, we use the `curve fit` function of the [scipy package](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html). This function takes, as input, `f` (the model funtion, or the curve to be fitted), `xdata` and `ydata`. Its outputs are `popt` and `pcov`. The former is the optimized parameter array (`a`, `b` and `omega`) via the least squares method, while latter is the estimated covariance matrix of said parameters.

```python
popt, pcov = curve_fit(f = cos, xdata = x, ydata = y)
```

Let's plot the data and the fit

```python
plt.plot(x, y, '.', label='Noisy data')
plt.plot(x, cos(x, *popt), ls='-', label='Fitted cosine')
plt.legend()
plt.title("The fitted model gives a = {a:.2f}, b = {b:.2f}, $\omega$ = {omega:.2f}".format(a = popt[0], b = popt[1], omega = popt[2])) 
```

Note that despite having a very noisy dataset, we the hit angular frequency dead on and amplitude and offset are very close up to the second decimal place. Feel free to tweak the parameters and functions on this notebook if you want. This is the quickest way to fit a dataset with a user designated model, but you can also give initial guesses to your parameters, set limits for them and even change the method for optimization. [Check the scipy docs if you're interested](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html).
